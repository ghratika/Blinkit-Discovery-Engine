# Phase 2: Data Exploration

Exploratory Data Analysis of the collected and normalized Blinkit reviews.

Run `run_collection.py` first to generate `data/processed/normalized.json`.

```python
import json
import pandas as pd
import sys
sys.path.insert(0, '..')

with open('../data/processed/normalized.json', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)
print(f'Total reviews: {len(df)}')
df.head()
```

```python
# Source distribution
print('Source distribution:')
print(df['source'].value_counts())
print()
print('Rating distribution:')
print(df['rating'].value_counts().sort_index())
```

```python
# Review length distribution
df['word_count'] = df['text'].str.split().str.len()
print('Review word count stats:')
print(df['word_count'].describe())
print()
print(f'Reviews with <= 10 words: {(df["word_count"] <= 10).sum()}')
print(f'Reviews with > 50 words: {(df["word_count"] > 50).sum()}')
```

```python
# Timestamp distribution
df['timestamp'] = pd.to_datetime(df['timestamp'], utc=True)
df['month'] = df['timestamp'].dt.to_period('M')
print('Reviews by month:')
print(df['month'].value_counts().sort_index())
```

```python
# Sample reviews from each source
for source in df['source'].unique():
    print(f'\n--- {source} ---')
    sample = df[df['source'] == source]['text'].head(3).tolist()
    for i, t in enumerate(sample, 1):
        print(f'{i}. {t[:120]}...')
```
