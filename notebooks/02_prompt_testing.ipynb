# Phase 3: Prompt Testing

This notebook is used to manually validate the Groq API extraction prompts against a few sample reviews before running the full `enrichment_runner.py` pipeline.

```python
import os
import json
from groq import Groq
from dotenv import load_dotenv

load_dotenv('../.env')
client = Groq(api_key=os.environ.get('GROQ_API_KEY'))
```

```python
from src.enrichment.prompts import ENRICHMENT_SYSTEM_PROMPT, ENRICHMENT_USER_PROMPT_TEMPLATE

sample_review = """I use Blinkit everyday for milk and bread. 
But I tried ordering cat food yesterday and it was completely expired! 
I'm a pet parent and this is unacceptable. Please improve quality checks."""

prompt = ENRICHMENT_USER_PROMPT_TEMPLATE.format(review_text=sample_review)

completion = client.chat.completions.create(
    messages=[
        {"role": "system", "content": ENRICHMENT_SYSTEM_PROMPT},
        {"role": "user", "content": prompt}
    ],
    model="llama3-8b-8192",
    temperature=0.0,
)

print(completion.choices[0].message.content)
```
